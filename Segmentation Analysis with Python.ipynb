#### Principal Component Analysis for exploring data
PCA transforms a multivariate data set containing metric variables to a new data set with variables– referred to as principal components– which are uncorrelated and ordered by importance. </br>
For EDA purpose, using PCA to create perceptual map. A perceptual map offers initial insights into how attributes are rated by respondents and, importantly, which attributes tend to be rated in the same way.</br>
<i>PCA chuyển đổi một tập dữ liệu đa biến chứa các biến định lượng thành một tập dữ liệu mới với các biến – được gọi là các Pricipal component – không tương quan và được sắp xếp theo mức độ quan trọng. </br>
Với mục đích thăm dò dữ liệu (EDA), PCA được sử dụng để tạo ra Perceptual map. Perceptual map cung cấp những hiểu biết ban đầu về cách các thuộc tính (ỏ đây là nhu cầu khi đi du lịch) được du khách đánh giá và, quan trọng hơn, những thuộc tính nào có xu hướng được đánh giá tương đồng nhau</br>
</i>

Since the data for segmentation variables are binary, it's reasonable to <b> not perform scaling.</b> in PCA. </br>
<i>Do dữ liệu là nhị phân, chúng ta sẽ <b>không thực hiện Scaling</b>  </i>

```python
def get_pca_summary_matrix(df, n_components=None, scale=False):
    pca = PCA(n_components=n_components)
    pca.fit(df)

    explained_variance = pca.explained_variance_
    explained_variance_ratio = pca.explained_variance_ratio_
    cumulative_proportion = np.cumsum(explained_variance_ratio)

    num_components = pca.n_components_
    columns = [f'PC {i+1}' for i in range(num_components)]
    index = ['Standard deviation', 'Proportion of Variance', 'Cumulative Proportion']

    summary_matrix = pd.DataFrame(index=index, columns=columns)
    summary_matrix.loc['Standard deviation'] = np.sqrt(explained_variance)
    summary_matrix.loc['Proportion of Variance'] = explained_variance_ratio
    summary_matrix.loc['Cumulative Proportion'] = cumulative_proportion

    return summary_matrix

df = df_transformed
pca_summary = get_pca_summary_matrix(df)
if pca_summary is not None:
    print("PCA Summary Matrix:")
    print(pca_summary)
```

The fact that the first few principal components do not explain much of the variance indicates that all the original items (survey questions) are needed as segmentation variables. They are not redundant. They all contribute valuable information<br> 
<i>Việc một vài thành phần chính đầu tiên không giải thích được nhiều phương sai cho thấy rằng tất cả các biến ban đầu của khảo sát đều cần thiết để thực hiện phân khúc. Tất cả biến này đều đóng góp thông tin có giá trị </i>

```python
def get_pca_rotation_matrix(df, n_components=4, scale=False):
    pca = PCA(n_components=n_components)
    pca.fit(df)
    rotation_matrix = pd.DataFrame(pca.components_.T,
                                   index=df.columns,
                                   columns=[f'PC {i+1}' for i in range(n_components)])
    return rotation_matrix

df = df_transformed

# Get the rotation matrix for the first 3 principal components
rotation = get_pca_rotation_matrix(df)
if rotation is not None:
    print("Rotation Matrix (Loadings)")
    print(rotation)
```

The loadings indicates how features project to the PC. Perceptual map is a way to understand how these motives are perceived relative to each other in a lower-dimensional space defined by the most important underlying dimensions of variance. Inspecting the rotation matrix reveals that the first principal component does not differentiate well between motives because all motives load on it positively. Principal components 2 and 3 display a more differentiated loading pattern of motives. We therefore use principal components 2 and 3 to create a perceptual map <br>
<i>Giá trị loadings cho thấy phương hướng và độ mạnh của các Features khi chiếu lên Pricipal component. Từ bảng trên có thể thấy được Pricipal component thứ nhất không làm rõ được sự khác biệt giữa các Features vì các giá trị loading chỉ có một hướng (+). Principal component thứ 2 và 3 thể hiện được nhiều sự khác biệt giữa các Features, vì vậy, tôi sẽ sử dụng 2 đại lượng này cho Perceptual map </i>

```python
pca = PCA(n_components=4)
X_pca = pca.fit_transform(df_transformed)

PC2 = X_pca[:, 1]
PC3 = X_pca[:, 2]
    
plt.figure(figsize=(10, 10))
plt.scatter(PC2, PC3, color='grey', s=8, alpha=0.5)

arrow_scale = 1.0 
pc_x = 'PC 2'
pc_y = 'PC 3'

for feature in rotation.index:
    x = rotation.loc[feature, pc_x] * arrow_scale
    y = rotation.loc[feature, pc_y] * arrow_scale
    plt.arrow(0, 0, x, y, head_width=0.03, head_length=0.05, fc='red', ec='red', alpha=0.7)

    ha = 'left' if x >= 0 else 'right'
    va = 'bottom' if y >= 0 else 'top'

    plt.text(x + 0.05 * np.sign(x), y + 0.05 * np.sign(y), feature,
             fontsize=10, va=va, ha=ha)

plt.axvline(0, color='black', linestyle='--', linewidth=0.8, alpha=0.5)
plt.axhline(0, color='black', linestyle='--', linewidth=0.8, alpha=0.5)

plt.xlabel('PC 2')
plt.ylabel('PC 3')
plt.title('Perceptual Map (PC 2 vs PC 3)')
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.gca().set_aspect('equal', adjustable='box')

plt.tight_layout()
plt.show()
```

```python
def scree_plot_kmodes(data, max_k, n_init, random_state):
    cost = []
    for num_clusters in range(1, max_k + 1):
        kmodes = KModes(n_clusters=num_clusters, init='Cao', n_init=n_init, random_state=random_state, verbose=0)
        kmodes.fit(data)
        cost.append(kmodes.cost_)

    plt.figure(figsize=(10, 6))
    plt.plot(range(1, max_k + 1), cost, marker='o')
    plt.title('Scree Plot for K-Modes')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Cost (Total Dissimilarity)')
    plt.xticks(range(1, max_k + 1))
    plt.grid(True)
    plt.show()
 
if __name__ == "__main__":
    data = df_kmodes
    scree_plot_kmodes(data, max_k=9, n_init=10, random_state=42)
```

The ARI score ranges from -1 to 1, where 1 indicates perfect agreement, 0 suggests the agreement is no better than random, and negative values imply the clusterings are less similar than expected by chance. Therefore, when comparing between clustering solutions (different k), we will look for the point right before the changing in number of cluster makes a big gap to the next one.
<br><i>Một sự sụt giảm rõ rệt về độ không tương đồng (Dissimilarity) được quan sát giữa phương án phân cụm 4 và 5. Sau đó, chỉ số Cost trở nên ổn định, cho thấy việc tăng số lượng cụm vượt quá 5 chỉ mang lại cải thiện rất nhỏ trong việc giảm độ không tương đồng. Do đó, dựa theo phương pháp “elbow”, 5 được xem là số cụm tối ưu.</i>  
<i>Tuy nhiên, hãy phân tích thêm để đánh giá tính ổn định của phương án này.</i>  
<i>Lần này, tôi sử dụng <b>Average Rand Index</b> làm chỉ số đo lường:</i>  
<i>Average Rand Index (ARI) là một chỉ số dùng để đánh giá mức độ tương đồng giữa các phương án phân cụm khác nhau trên cùng một tập dữ liệu, thường được sử dụng trong bối cảnh phân tích tính ổn định. Chỉ số ARI có giá trị từ -1 đến 1, trong đó 1 biểu thị sự đồng thuận hoàn hảo, 0 cho thấy mức độ đồng thuận không tốt hơn ngẫu nhiên, và giá trị âm cho thấy các phân cụm kém tương đồng hơn cả kỳ vọng ngẫu nhiên. Do đó, khi so sánh giữa các phương án phân cụm (với số lượng cụm k khác nhau), chúng ta sẽ tìm điểm ngay trước khi sự thay đổi về số cụm tạo ra một khoảng cách lớn so với phương án tiếp theo.</i>

```python
data = df_kmodes.values

n_clusters_range = range(2, 10)
n_bootstraps = 200
n_init = 10
random_seed = 42
if random_seed is not None:
    np.random.seed(random_seed)

stability_scores = defaultdict(list)

for n_clusters in n_clusters_range:
    print(f"Analyzing stability for {n_clusters} clusters...")
    for i in range(n_bootstraps):
        # Create a bootstrap sample
        bootstrap_sample_indices = resample(np.arange(data.shape[0]), replace=True, random_state=np.random.randint(0, 10000))
        bootstrap_sample = data[bootstrap_sample_indices]

        # Run K-modes on the bootstrap sample
        km_bootstrap = KModes(n_clusters=n_clusters, init='Cao', n_init=n_init, random_state=np.random.randint(0, 10000), verbose=0)
        labels_bootstrap = km_bootstrap.fit_predict(bootstrap_sample)

        # Run K-modes on the original data
        km_original = KModes(n_clusters=n_clusters, init='Cao', n_init=n_init, random_state=random_seed, verbose=0)
        labels_original = km_original.fit_predict(data)

        # Map bootstrap labels back to the original data indices
        mapped_bootstrap_labels = np.full(data.shape[0], -1)
        for j, original_index in enumerate(bootstrap_sample_indices):
            mapped_bootstrap_labels[original_index] = labels_bootstrap[j]

        # Consider only the samples present in the bootstrap
        indices_in_bootstrap = np.where(mapped_bootstrap_labels != -1)[0]
        original_labels_subset = labels_original[indices_in_bootstrap]
        bootstrap_labels_subset = mapped_bootstrap_labels[indices_in_bootstrap]

        # Calculate ARI
        if len(np.unique(original_labels_subset)) > 1 and len(np.unique(bootstrap_labels_subset)) > 1:
            ari = adjusted_rand_score(original_labels_subset, bootstrap_labels_subset)
            stability_scores[n_clusters].append(ari)

# Calculate average ARI for each number of clusters
average_ari = {k: np.mean(v) for k, v in stability_scores.items()}
ari_df = pd.DataFrame.from_dict(average_ari, orient='index', columns=['Average ARI'])
ari_df = ari_df.sort_index()

plt.figure(figsize=(10, 6))
plt.plot(ari_df.index, ari_df['Average ARI'], marker='o')
plt.title('Stability Analysis of K-Modes using Bootstrapping')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Average Adjusted Rand Index (ARI)')
plt.xticks(n_clusters_range)
plt.grid(True)
plt.ylim(-0.1, 1.1) # while ARI is limited from -1 to 1, setting 1.1 is for better visual
plt.show()
```

The 5-cluster solution, or even other solutions, shows moderate level (0.4-0.6), revealing a sensitivity of cluster solutions across initiation and indicating that its clustering is only marginally better than random assignment. </br>

#### 5.1.2 Bagged clustering
For this model, I will use bagged clustering to improve the sensitivity and combind with hierarchical clustering as we can't pick out the most optimal number of cluster
<br><i>Với mô hình này, tôi sẽ sử dụng phương pháp **bagged clustering** nhằm cải thiện độ nhạy, kết hợp với **hierarchical clustering**, vì chúng ta không thể xác định rõ ràng số lượng cụm tối ưu nhất.</i>

<b>Step 1: Implement Bagged K-modes </b><br>
We will choose number of cluster relative higher than the number of clusters expected to exist in the data. Leisch (1999) shows that if k is larger than necessary, segments artificially split up in this step are merged during hierarchical clustering. This is one of the reasons why we should use bagged clustering when we're not sure about choosing the number of clusters.
<br><i>Chúng ta sẽ chọn số lượng cụm cao hơn so với số cụm được kỳ vọng thực sự có trong dữ liệu. Theo Leisch (1999), nếu giá trị k được chọn lớn hơn mức cần thiết, các phân khúc bị chia tách một cách có ý định ở bước này sẽ được gộp lại trong quá trình phân cụm phân cấp (hierarchical clustering). Đây là một trong những lý do vì sao chúng ta nên sử dụng phương pháp bagged clustering khi chưa chắc chắn về việc lựa chọn số lượng cụm.</i>

```python
df_bagged = df_transformed.copy()
df_bagged.head()
```

```python
def bagged_kmodes(data, n_clusters, n_bags=100, n_init=5, random_state=None):
    if random_state is not None:
        np.random.seed(random_state)
    all_bag_labels = []
    n_samples = data.shape[0]
    for i in range(n_bags):
        bootstrap_indices = resample(np.arange(n_samples), replace=True, random_state=np.random.randint(0, 10000))
        bootstrap_sample = data.iloc[bootstrap_indices].values
        km = KModes(n_clusters=n_clusters, init='Cao', n_init=n_init, random_state=np.random.randint(0, 10000), verbose=0)
        labels = km.fit_predict(bootstrap_sample)
        # Map bootstrap labels back to original indices
        original_labels = np.full(n_samples, -1)
        for j, original_index in enumerate(bootstrap_indices):
            original_labels[original_index] = labels[j]
        all_bag_labels.append(original_labels)
    return all_bag_labels

if __name__ == "__main__":
    data = df_bagged
    n_clusters = 8
    n_bags = 5  # Use a smaller number for testing
    bagged_labels = bagged_kmodes(data, n_clusters, n_bags=n_bags, random_state=42)

    print("Bagged K-modes labels for each bag (first 5):")
    for i in range(min(5, len(bagged_labels))):
        print(f"Bag {i+1}: {bagged_labels[i][:10]}")
```

<b>Step 2: Calculate the Co-occurrence Matrix </b>

```python
def calculate_co_occurrence(all_bag_labels, n_clusters):
    n_samples = len(all_bag_labels[0])
    co_occurrence_matrix = np.zeros((n_samples, n_samples))
    n_bags = len(all_bag_labels)

    for i in range(n_samples):
        for j in range(i + 1, n_samples):
            count = 0
            for bag_labels in all_bag_labels:
                if bag_labels[i] != -1 and bag_labels[j] != -1 and bag_labels[i] == bag_labels[j]:
                    count += 1
            co_occurrence_matrix[i, j] = count / n_bags
            co_occurrence_matrix[j, i] = count / n_bags
    return co_occurrence_matrix

if __name__ == "__main__":
    n_clusters = 3  
    co_occurrence_matrix = calculate_co_occurrence(bagged_labels, n_clusters)
    print("\nCo-occurrence matrix (first 5x5):")
    print(co_occurrence_matrix[:5, :5])
```

<b>Step 3: Perform Hierarchical Clustering on the Co-occurrence Matrix</b>

```python
def perform_hierarchical_clustering(co_occurrence_matrix, method='ward'):
    # Convert co-occurrence to a distance matrix (1 - similarity)
    distance_matrix = 1 - co_occurrence_matrix
    # Ensure diagonal is 0 (redundant but good practice)
    np.fill_diagonal(distance_matrix, 0)

    # Convert the square distance matrix to condensed form
    condensed_distance_matrix = distance.pdist(distance_matrix, metric='euclidean') # Using 'euclidean' as the metric for linkage

    linked = linkage(condensed_distance_matrix, method=method, optimal_ordering=True)
    return linked

if __name__ == "__main__":
    linked_matrix = perform_hierarchical_clustering(co_occurrence_matrix, method='ward')
    print("\nLinkage matrix (first 5 rows):")
    print(linked_matrix[:5])
```

<b>Step 4: Visualize the Dendrogram and Choose the Number of Clusters </b>

```python
def visualize_dendrogram(linked):
    plt.figure(figsize=(12, 6))
    dendrogram(linked, orientation='top')
    plt.title('Hierarchical Clustering of Bagged K-Modes Results')
    plt.xticks([])
    plt.xlabel('Data Points')
    plt.ylabel('Distance')
    plt.axhline(y=65, color='grey', linestyle='--',alpha=0.7)
    plt.show()

if __name__ == "__main__":
    visualize_dendrogram(linked_matrix)
```

Trong khi tôi vẫn đang tìm hiểu thêm về cách hoạt động của thuật toán **Neural Gas** (và sẽ tiếp tục khám phá cùng ChatGPT), hãy cùng cân nhắc những gợi ý từ các chuyên gia và tiến hành xây dựng thuật toán.</i>

```python
df_neutral = df_transformed.copy()
df_neutral.head()
```

```python
feature_columns = df_neutral.columns  
X = df_neutral[feature_columns].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

def neural_gas(X, n_clusters=6, epochs=100, epsilon=0.3, lambda0=3, random_state=42):
    np.random.seed(random_state)
    n_samples, n_features = X.shape

    # Initialize prototypes deterministically
    prototypes = X[np.random.choice(n_samples, n_clusters, replace=False)].astype(np.float64)

    # Fixed training order (no shuffling)
    indices = np.arange(n_samples)

    for epoch in range(epochs):
        for i in indices:
            x = X[i]
            distances = np.linalg.norm(prototypes - x, axis=1)
            ranking = np.argsort(distances)
            for rank, index in enumerate(ranking):
                influence = np.exp(-rank / lambda0)
                prototypes[index] += epsilon * influence * (x - prototypes[index])

    # Assign each sample to closest prototype
    labels = np.argmin(np.linalg.norm(X[:, None, :] - prototypes[None, :, :], axis=2), axis=1)
    return labels + 1 
if __name__ == "__main__":
    labels_neutral = neural_gas(X_scaled, n_clusters=6, epochs=100, random_state=42)
    print(labels_neutral[:20])
```

<b>Please note:</b> Although the clustering algorithm is deterministic in how it groups similar tourists together, the numeric labels assigned to each cluster (e.g., Cluster 1, Cluster 2) are arbitrary and may change order across runs. This means that while the group compositions remain the same, the cluster numbers might not match exactly from one run to another. To ensure consistency in presentation, we can align or rename clusters based on their characteristics rather than relying on label numbers.
<br><i><b>Lưu ý:</b> Các nhãn số được gán cho mỗi cụm (ví dụ: Cụm 1, Cụm 2) là tùy ý và có thể thay đổi thứ tự qua các lần chạy.
```python
import seaborn as sns
from scipy.stats import chi2_contingency

# Function to compute Pearson residuals
def compute_residuals(df_discriptor, cat_var, segment_col='Segment'):
    contingency = pd.crosstab(df_discriptor[cat_var], df_discriptor[segment_col])
    chi2, _, _, expected = chi2_contingency(contingency)
    observed = contingency.values
    residuals = (observed - expected) / np.sqrt(expected)
    residuals_df = pd.DataFrame(residuals, index=contingency.index, columns=contingency.columns)
    return residuals_df

cat_vars = ['Gender', 'Education','Income', 'Occupation', 'State','Relationship.Status']

n_vars = len(cat_vars)
fig, axes = plt.subplots(n_vars, 1, figsize=(12, n_vars * 4.5))
order_income = ['Less than $30,000','$30,001 to $60,000', '$60,001 to $90,000','$90,001 to $120,000','$120,001 to $150,000','$150,001 to $180,000','$180,001 to $210,000','$210,001 to $240,000','more than $240,001']
# Generate heatmap for each variable
for i, var in enumerate(cat_vars):
    ax = axes[i]
    res_df = compute_residuals(df_discriptor, var)
    if var == 'Income':
        res_df = res_df.reindex(order_income)
    sns.heatmap(res_df, annot=True, fmt=".2f", cmap="RdBu",vmin = -2, vmax = 2, center=0, cbar=i == 0, ax=ax)   
    ax.set_title(f"Pearson Residuals for {var}", fontsize=12)
    ax.set_xlabel("")
    ax.set_ylabel("")
    
plt.tight_layout()
plt.show()
```

Characteristics with residual values of 2 or higher are considered noticely representative of the segment:<br>
Desciption of 4 focused segments:</br>
- Segment 2: Predominantly features individuals with an education level of 4 and those who are retired.</br>
- Segment 3: Is characterized by tradespeople, often residing in NSW, with an average income range of AUD 120,000 to 150,000.</br>
- Segment 4: Largely comprises individuals who are widowed.</br>
- Segment 6: Is defined by professionals living in VIC with a partner, exhibiting a high level of education and an average income between AUD 90,000 and 120,000.</br>

#### 7.2  Metric Descriptor Variables
We have: 'Age', 'Obligation', 'NEP', 'Vacation.Behaviour', 'Education'<br>
To see if there's any significant difference in the distribution of these variables across segments, Tukey HDS chart is suggested here
<br><i>Để xem liệu có bất kỳ sự khác biệt đáng kể nào trong phân phối của các biến này giữa các phân khúc hay không, biểu đồ Tukey HSD được đề xuất ở đây

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import itertools

numeric_vars = ['Age', 'Obligation', 'NEP', 'Vacation.Behaviour']

# Loop through each variable
for var in numeric_vars:
    # Run Tukey HSD
    tukey = pairwise_tukeyhsd(endog=df_discriptor[var], groups=df_discriptor['Segment'], alpha=0.05)
    summary_df = pd.DataFrame(data=tukey._results_table.data[1:], columns=tukey._results_table.data[0])
    
    plt.figure(figsize=(8, 5))
    ax = sns.boxplot(data=df_discriptor, x='Segment', y=var, hue='Segment', palette='Set2', legend=False)
    plt.title(f'{var} by Segment (Tukey HSD)')
    
    y_max = df_discriptor[var].max()
    y_min = df_discriptor[var].min()
    h = (y_max - y_min) * 0.05  # height between lines
    step = 0
    
    for i, row in summary_df.iterrows():
        if row['reject']:
            group1, group2 = row['group1'], row['group2']
            x_labels = df_discriptor['Segment'].unique().tolist()
            x1 = x_labels.index(group1)
            x2 = x_labels.index(group2)
            y = y_max + h * step
            step += 1
          
            ax.plot([x1, x1, x2, x2], [y, y + h/2, y + h/2, y], lw=1.5, color='black')
            ax.text((x1 + x2) / 2, y + h/2 + 0.01 * y_max, '*', ha='center', va='bottom', fontsize=12)
    
    plt.tight_layout()
    plt.show()
```

The line connects the two box plots indicates the significan difference between them <br>
Compared to Segment 3 and 4, it's not suprisingly when Segment 2 is significantly higher in NEP score and moral obligation to behave environmentally friendly. And the segment 3 has higher NEP score than segment 6.<br>
